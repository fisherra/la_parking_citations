### Read in the Dataset

```python
la_ticket_full = pd.read_csv("~/Documents/data_science/py_la_tickets/parking_citation.csv")
```

### Index Dataset for 2017 Results Only

```python
la_ticket_issue = la_ticket_full['Issue Date']
la_ticket_issue.tail()
```

```python
type(la_ticket_issue)
```

```python
sum(la_ticket_issue.str.contains('2017') == True)
```

```python
la_ticket_2017_index = la_ticket_issue.str.contains('2017')
```

```python
la_ticket_2017 = la_ticket_full[la_ticket_2017_index]
```

```python
len(la_ticket_2017)
```

### Remove 'Useless' Variables

```python
la_ticket_2017 = la_ticket_2017.drop(['Ticket number', 'VIN', 'Marked Time', 'Plate Expiry Date', 'Route', 'Agency'], axis=1)
```

```python
la_ticket_2017.head()
```

### Split Issue Date into Month / Day

```python
date_split = la_ticket_2017['Issue Date'].str.split('/', n=3, expand=True)
date_split.head()
#new = data["Name"].str.split(" ", n = 1, expand = True) 
```

```python
# making seperate first name column from new data frame 
la_ticket_2017["Month"]= date_split[1] 
la_ticket_2017['Split_2'] = date_split[2]

la_ticket_2017.head()
```

```python
second_split = la_ticket_2017['Split_2'].str.split(' ', n=1, expand = True)
second_split.head()
```

```python
la_ticket_2017["Day"]= second_split[0] 
la_ticket_2017.head()
```

### Write Reduced 2017 Dataset

```python
la_ticket_2017.to_csv('la_ticket_2017.csv')
```

### Divide Dataset into 3 parts for github upload

```python
seq_1 = ['01','02','03','04']
seq_2 = ['05','06','07','08']
seq_3 = ['09','10','11','12']
```

```python
index_1 = la_ticket_2017['Month'].isin(seq_1)
index_2 = la_ticket_2017['Month'].isin(seq_2)
index_3 = la_ticket_2017['Month'].isin(seq_3)
```

```python
la_ticket_2017_1 = la_ticket_2017[index_1]
la_ticket_2017_2 = la_ticket_2017[index_2]
la_ticket_2017_3 = la_ticket_2017[index_3]
```

```python
la_ticket_2017_1.to_csv('la_ticket_2017_1.csv')
la_ticket_2017_2.to_csv('la_ticket_2017_2.csv')
la_ticket_2017_3.to_csv('la_ticket_2017_3.csv')
```
